```python
# Build chatbot dataset from Excel and intent_responses
chat_data = []
for _, row in df.iterrows():
    intent = row["Intent"].strip().lower()
    answer = intent_responses.get(intent, "Sorry, I don’t have an answer for that intent.")
    chat_data.append({
        "question": row["Question"].strip().lower(),
        "intent": intent,
        "answer": answer
    })
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Extract questions for vectorization
questions = [item['question'] for item in chat_data]
vectorizer = TfidfVectorizer()
question_vectors = vectorizer.fit_transform(questions)

# Chatbot function
def chatbot_response(user_input):
    user_input = user_input.lower()
    user_vector = vectorizer.transform([user_input])

    similarity = cosine_similarity(user_vector, question_vectors)
    idx = similarity.argmax()
    score = similarity[0][idx]

    if score > 0.3:
        response = chat_data[idx]
        return f"Bot: {response['answer']}"
    else:
        return "Bot: Sorry, I didn't understand that. Can you rephrase?"
```
